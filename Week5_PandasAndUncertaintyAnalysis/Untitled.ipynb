```python
import numpy as np
import pandas as pd
import scipy.optimize as scopt
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("MadeUpData.xlsx",sheet_name = "Data")
df = df.dropna()
```

```python
df
```

```python
def curve (x, a0, a1, a2, a3):
    y = a0 + a1*np.tanh(a2*(x-a3))
    return y
def get_dely (x, afit, c):
    F = []
    y = curve(x, *afit)
    for i in range(len(afit)):
        afitnew = afit + 0.0
        ai = afitnew[i]
        newai = (1 + 0.01)*ai
        afitnew[i] = newai
        ynew = curve(x, *afitnew)
        F.append((ynew - y)/(newai - ai))
    F=np.array(F)
    dely2 = F.dot(c.dot(F))
    print(F)
    return np.sqrt(dely2)
afit, c = scopt.curve_fit(curve, df["x"],df["y"],sigma = df["dely"])
```

```python
x = np.linspace (min(df["x"]),max(df["x"]),1000)
y = curve(x,*afit)
plt.scatter(df["x"],df["y"],s = 10)
dely = np.array([get_dely(xx,afit,c) for xx in x])
yplus = (y+dely)
yminus = (y-dely)
plt.fill_between(x, yminus, y2 = yplus, color = 'red')
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Y vs X")
#dely
```
